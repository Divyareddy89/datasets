```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import MinMaxScaler,LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection._split import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
t1=pd.read_csv("titanic.csv",index_col="PassengerId")
t1
```

```python
le_Pclass=LabelEncoder()
le_Sex=LabelEncoder()
le_Age=LabelEncoder()
le_Fare=LabelEncoder()
```

```python
t1["P_class"]=le_Pclass.fit_transform(t1["Pclass"])
t1["S_ex"]=le_Sex.fit_transform(t1["Sex"])
t1["A_ge"]=le_Age.fit_transform(t1["Age"])
t1["F_are"]=le_Fare.fit_transform(t1["Fare"])
t1
```

```python
t2= t1.drop(["Pclass","Sex","Age","Fare"],axis="columns")


X=t1[["P_class","S_ex","A_ge","F_are"]]
y=t1["Survived"]
y
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```
